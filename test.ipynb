```python
import numpy as np
import pandas as pd

# Mit Seed: Damit jedes mal die gleichen Zahlen generiert werden (reproducability)
np.random.seed(42)

# Erzeugen der X und Y Koordinaten als Gauss'sche Verteilung mit Mittelwert (loc) von 0 und Standardabweichung (scale) von 1
x = np.random.normal(loc=0, scale=1, size=100)
y = np.random.normal(loc=0, scale=1, size=100)

# Zusammenführen der beiden Listen zu einem DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Scatter Plot der Punkte
df.plot(kind='scatter', x='x', y='y')
```
